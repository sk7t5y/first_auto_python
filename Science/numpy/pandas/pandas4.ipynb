```python
import pandas as pd

ages = [0,2,10,21,23,37,31,61,20,41,32,100]
bins = [0,15,25,35,60,99]
lb = ["미성년자", "청년", "중년", "장년", "노년"]

cats = pd.cut(ages, bins, labels=lb)
cats

ages_df = pd.DataFrame(ages, columns= ["ages"])
ages_df["age.cat"] = cats
ages_df
```

```python
ages_df["age.cat"].value_counts()
```
